```python

from gettext import install
from operator import index
from bs4 import BeautifulSoup as bs
import re
import urllib.request
import lxml.html
import urllib.request
import requests
import pandas as pd
import json
from time import sleep
# from settings import URL
URL = "https://www.imdb.com/india/top-rated-indian-movies/"
import uuid
uuid4 = uuid.uuid4()
class scraper:
        def __init__(self) -> None:
                # self.driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
                # self.url = self.driver.get(URL)
                request = requests.get(URL)
                htmlContent = request.content
                self.soup =bs(htmlContent, "html.parser")
              
                
        def nevigate_page(self):
                parent = self.soup.find('td', class_= 'titleColumn')
                child = parent.find('a')['href']
                print(child)
                        
                      
                
                """ sheet.append([self.rank, self.name, self.year, self.rating])
                excel.save('IMBD Movies Ratings.xlsx') """
                        
        def get_details(self):
                self.indian_movies = self.soup.find('tbody', class_='lister-list').find_all('tr')
                list = []
                for name in self.indian_movies:
                        # movies = list(zip(Name, Rank, Year, Rating, Link ))
                        # self.soup.find_all('h3')
                        # print(self.soup)
                        self.rank = name.find('td', class_='titleColumn').get_text(strip=True).split('.')[0]
                        self.name = name.find('td', class_='titleColumn').a.text
                        #self.link = show.find('td', 'href a') 
                        # self.term = name.find('span', class_='terms').get_text(strip=True).split('.')[0]
                        # print(self.name, self.term)
                        self.year = name.find('td', class_='titleColumn').span.text.strip('()')
                        self.rating = name.find('td', class_='ratingColumn imdbRating').strong.text
        
                        top_indian_movies = {
                        'Rank' : self.rank,
                        'Name' : self.name,
                        'Year' : self.year,
                        'Rating' : self.rating}
                        list.append(top_indian_movies)
                        # print(top_movies)
                        df = pd.DataFrame(list, columns = ['Rank', 'Name', 'Year', 'Rating'])
                        # df.reset_index(drop = True)
                        df.to_csv('imdb_top_250_Indian_movies.csv',index=False)
                        print(df)
                        
        def main(self) -> None:
                self.nevigate_page()
                self.get_details()
                #self.page_scroll()
                            
if __name__ == "__main__":
    DPS = scraper()
    DPS.main()

        
```
