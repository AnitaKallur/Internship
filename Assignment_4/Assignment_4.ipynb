Scrape the details of most viewed videos on YouTube from Wikipedia.

A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url = driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")

container = driver.find_elements(By.XPATH, value="//table[@class='wikitable sortable jquery-tablesorter']")
th_tag = []

heading = driver.find_element(By.TAG_NAME, "caption").text
details = driver.find_elements(By.TAG_NAME, "tbody")
for d in details:
    tr_tag = d.find_elements(By.TAG_NAME, 'tr')
    for r in tr_tag:
        th_tag.append(r.text.strip())
driver.close()





```

Scrape the details of trending repositories on Github.com. Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code.


```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = driver.get("https://github.com/")
# wait_time = sleep(10)
sleep(1)
driver.maximize_window()

input = driver.find_element(By.XPATH, '//input[@type="text"]')
input.send_keys("Trending Repository")
sleep(2)

element_click = driver.find_element(By.XPATH, "/html/body/div[1]/header/div/div[2]/div/div/div[1]/div/div/form/label/div/ul[3]/li[3]/a/div[2]")
element_click.click()
sleep(2)

container = driver.find_elements(By.XPATH, "//ul[@class='repo-list']")
repo_title = []
urls_repo = []
repo_description = []
for repo in container:
    try:
        
        url_container = driver.find_elements(By.XPATH, "//a[@class='v-align-middle']")
        for url in url_container:
            urls_repo.append(url.get_attribute("href"))
    except NoSuchElementException:
            urls_repo.append('-')
    
    try:
        title = repo.find_elements(By.XPATH, "//div[@class='f4 text-normal']")
        for t in title: 
            repo_title.append(t.text)
        
    except NoSuchElementException:
        repo_title.append('-')
        
    try:
        description  = repo.find_elements(By.XPATH, "//p[@class='mb-1']")
        for d in description: 
            repo_description.append(d.text)
        
    except NoSuchElementException:
            repo_description.append('-')
    
contributors_count = []
language = []
sleep(2)
for u in urls_repo:
    driver.get(u)
    sleep(3)
    try:
        for v in driver.find_elements(By.XPATH, "/html/body/div[4]/div/main/turbo-frame/div/div/div/div[3]/div[2]/div/div[5]/div/h2/a/span"):
            sleep(2)
            contributors_count.append(v.text)
    
    except NoSuchElementException:
            contributors_count.append('-')
        
    try:
        lang_used = driver.find_elements(By.XPATH, "//span[@class='color-fg-default text-bold mr-1']")
        sleep(2)
        for x in lang_used:    
            language.append(x.text)
    
    except NoSuchElementException:
            language.append('-')
    
    
```

```python

repo_details = {'Repo_Title': repo_title, 'Description': repo_description, 'Repo_Link': urls_repo, 'Contributors Count': contributors_count, 'Language Used': language}
df = pd.DataFrame.from_dict(repo_details, orient='index')
df = df.transpose()
df
```
