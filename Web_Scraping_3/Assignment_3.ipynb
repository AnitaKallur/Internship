1. Write a python program which searches all the product under a particular product from www.amazon.in.
The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search
for guitars.
2. In the above question, now scrape the following details of each product listed in first 3 pages of your
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search
results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Connecting and automating the webpage vis chrome browser
   
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url = driver.get("https://www.amazon.in")

# content = driver_web.page_source
# soup = BeautifulSoup(content)
try:
    
    popup = driver.find_element(by=By.XPATH, value="//button[@class='_2KpZ6l _2doB4z']")
    popup.click()
    
except:
    pass
driver.maximize_window()
driver.implicitly_wait(10)


item = driver.find_element(by=By.ID, value= 'twotabsearchtextbox')
sleep(2)
item.send_keys('laptop')
sleep(2)
search = driver.find_element(by=By.ID, value="nav-search-submit-button")
sleep(1)
search.click()
sleep(1)
# driver.quit()
```

```python
name_container = []
urls_laptop = []
price = []
delivery = []
# availability = []
start = 0
end = 3
for page in range(start, end):
    url_container = driver.find_elements(by=By.XPATH, value="//a[@class ='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
    for url in url_container:
        # for page in range(0,1,1):
       
        urls_laptop.append(url.get_attribute('href'))
        
            
    next_page = driver.find_element(by=By.XPATH, value="//a[@class ='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    next_page.click()
    sleep(2)
    
    container_laptop = driver.find_elements(by=By.XPATH, value="//div[@class='a-section a-spacing-small a-spacing-top-small']")

    for i in container_laptop:
        try: 
            for j in  driver.find_elements(by=By.XPATH, value="//span[@class='a-size-medium a-color-base a-text-normal']"):
                name_container.append(j.text)
        except NoSuchElementException:
                name_container.append('-')
                
        try:
            for k in driver.find_elements(by=By.CLASS_NAME, value="a-price-whole"):
                price.append(k.text)
        except NoSuchElementException:
                price.append('-')
                
        try:
            for l in driver.find_elements(by=By.XPATH, value="//div[@class='a-section a-spacing-none a-spacing-top-micro']"):
                delivery.append(l.text)
        except NoSuchElementException:
                delivery.append('-')
                
    
availability = [] 
for u in urls_laptop:
    driver.get(u)
    sleep(2)
    try:
        for m in driver.find_elements(by=By.XPATH, value="//div[@id = 'availability']"):
            availability.append(m.text)
    except NoSuchElementException:
            availability.append('-')
                
            
    

                
                
                
                
```

maps
