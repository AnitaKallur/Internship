```python

for z in range(0, 2):
    options = FirefoxOptions()
    driver = webdriver.Firefox(executable_path=GeckoDriverManager().install(), options=options)
    url = driver.get("https://www.kayak.co.uk/flights/LON-BLR/2023-03-06/2023-03-18?sort=bestflight_a")
    driver.maximize_window()
    sleep(1)
    try: 
        pop_up = driver.find_element(by=By.XPATH, value="//div[@aria-label='Close']").click()
        sleep(10)
    except:
        NoSuchElementException

    soup = BeautifulSoup(driver.page_source, 'html.parser')

    element = driver.find_element(by=By.TAG_NAME, value="body")
    page = driver.find_elements(by=By.XPATH, value="//div[@class='Base-Results-ResultsList Flights-Results-FlightResultsList']")
    start = 0
    end = 10
    for page in range(start,end):
        while True:
            element.send_keys(Keys.PAGE_DOWN)
            sleep(3)
            try:
                driver.find_element(by=By.XPATH, value= "//div[@class='resultsPaginator']").click()
                sleep(4)
                
            except:
                NoSuchElementException
                
        
            # break
            try:
                driver.find_element(by=By.XPATH, value="//span[@class='icon']").click()
                sleep(2)
            except:
                NoSuchElementException
                break
        
        # for y in range(3):
            airline = []
            airlines = soup.find_all('div', class_='J0g6-operator-text',text=True)
            for i in airlines:       
                airline.append(i.text)
                
                        
            # def get_total_stops(soup):
            stops_list = []
            stops = soup.find_all('div',class_='vmXl vmXl-mod-variant-default')
            for i in stops:  
                for j in i.find_all('span',class_='JWEO-stops-text'):      
                  stops_list.append(j.text)
            # return stops_list


            departure = []
            depart = soup.find_all('div',class_='c_cgF c_cgF-mod-variant-full-airport')
            for i in depart:            
                for j in i.find_all('div',class_='EFvI'):                
                    departure.append(j.text)
            
    # def get_price(soup):
            prices = []
            price = soup.find_all('div',class_='f8F1-above')
            for i in price:            
                for j in i.find_all('div', class_='f8F1-price-text'):                  
                    prices.append(j.text)
                # return prices

            # def get_duration(soup):
            duration_list = []
            duration = soup.find_all('div' , class_='xdW8 xdW8-mod-full-airport')
            for i in duration:            
                for j in i.find_all('div',class_='vmXl vmXl-mod-variant-default'):               
                    duration_list.append(j.text)
                    break
        
            columns_list = ['airline', 'stops_list', 'departure', 'prices', 'duration_list']
            kayak_list = ({'Airline': airline, 'stops_list': stops_list, 'Departure': departure, 'Prices': prices, 'Duration': duration_list})
            # df = pd.DataFrame.from_dict(kayak_list, columns= ['airline', 'stops_list', 'departure', 'prices', 'duration_list'])
            
            # page+=1    
            
            df = pd.DataFrame.from_dict(kayak_list, orient='index')
            # df.columns = ['airline', 'stops_list', 'departure', 'prices', 'duration_list']
            
            # df.columns = ['airline', 'stops_list', 'departure', 'prices', 'duration_list']
            # df.columns = columns_list
            df= df.transpose()
            # df.to_csv('kayak_data.csv')
            if os.path.exists('kayak_data_1.csv'):
                df.to_csv('kayak_data_1.csv',mode='a', index=False, quoting = False, index_label=None, header = True,line_terminator='\n', quotechar=' ')
                df.to_excel(r'/Users/prabhuswamikallur/Desktop/kayak_data.xlsx', index=False, index_label=None  , header=True)
            else:
                df.to_csv('kayak_data_1.csv', mode='a', index=False, quoting = False, index_label=None, header = True,line_terminator='\n', quotechar=' ')
            
            print(df)
          
    
```
